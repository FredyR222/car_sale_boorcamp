```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulación de datos
data = {'Genero': ['Masculino', 'Femenino', 'Masculino', 'Femenino', 'Masculino', 'Femenino'],
        'Ingreso': [50000, 52000, 48000, 55000, 53000, 58000]}
df = pd.DataFrame(data)

# Agrupar por género y calcular ingresos promedio
ingresos_por_genero = df.groupby('Genero')['Ingreso'].mean()

# Crear el gráfico de barras
plt.figure(figsize=(8,5))
ingresos_por_genero.plot(kind='bar', color=['blue', 'pink'])
plt.title('Ingresos Promedio por Género')
plt.xlabel('Género')
plt.ylabel('Ingreso Promedio')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
